## Q1. What is hypothesis testing in statistics?

### Answer:  
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It helps determine whether the observed data provides enough evidence to reject a specific assumption about a population parameter.

In hypothesis testing, two competing statements are formed:  

- **Null Hypothesis (H‚ÇÄ):** Assumes that there is no effect, no difference, or no relationship. It represents the default or status quo.  
- **Alternative Hypothesis (H‚ÇÅ or Ha):** Suggests that there is an effect, a difference, or a relationship. It represents what the researcher wants to prove.  

## Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

### Answer:  
In hypothesis testing, two statements are made to test assumptions about a population:

- **Null Hypothesis (H‚ÇÄ):**  
  It states that there is **no effect**, **no difference**, or **no relationship**. It represents the current situation or a claim that is assumed to be true unless there is strong evidence against it.  

- **Alternative Hypothesis (H‚ÇÅ or Ha):**  
  It suggests that there **is an effect**, **a difference**, or **a relationship**. Researchers try to find evidence to support this claim.  

### Difference:  
The null hypothesis represents the default or no-change condition, while the alternative hypothesis proposes that something has changed or there is a significant effect. Hypothesis testing helps determine which statement is supported by the data.


## Q3. What is the significance level in hypothesis testing, and why is it important?

### Answer:  
The **significance level (Œ±)** is the probability of making a **Type I error**, which means rejecting the null hypothesis when it is actually true. It defines how much risk of error the researcher is willing to accept.

### Common Values of Œ±:  
- 0.05 (5%) ‚Äî Most commonly used.  
- 0.01 (1%) ‚Äî Stricter, used for more precise studies.  

### Importance:  
- It helps decide whether the observed results are **statistically significant**.  
- A smaller Œ± reduces the chances of making incorrect conclusions.  
- It provides a clear rule for decision-making in hypothesis testing.  

## Q4. What does a P-value represent in hypothesis testing?

### Answer:  
The **P-value** is the probability of observing the sample results, or more extreme results, if the **null hypothesis (H‚ÇÄ)** is true.

### Interpretation:  
- A **small P-value** (typically ‚â§ 0.05) indicates strong evidence **against** the null hypothesis, so we reject H‚ÇÄ.  
- A **large P-value** (> 0.05) indicates weak evidence against H‚ÇÄ, so we fail to reject H‚ÇÄ.  

## Q5. How do you interpret the P-value in hypothesis testing?

### Answer:  
The **P-value** helps us decide whether to reject or fail to reject the null hypothesis (H‚ÇÄ).

### Interpretation Rules:  
- If **P-value ‚â§ significance level (Œ±)** (commonly 0.05):  
  ‚Üí Reject the null hypothesis.  
  ‚Üí The result is **statistically significant**.  

- If **P-value > significance level (Œ±)**:  
  ‚Üí Fail to reject the null hypothesis.  
  ‚Üí There is **not enough evidence** to say the result is significant.  

### Example:  
- If P-value = 0.03 and Œ± = 0.05 ‚Üí Reject H‚ÇÄ, result is significant.  
- If P-value = 0.08 and Œ± = 0.05 ‚Üí Fail to reject H‚ÇÄ, result is not significant.  

## Q6. What are Type 1 and Type 2 errors in hypothesis testing?

### Answer:  
In hypothesis testing, two types of errors can occur when making decisions:

### Type 1 Error (False Positive):  
- Occurs when the **null hypothesis (H‚ÇÄ)** is rejected even though it is actually true.  
- In simple terms: Detecting an effect or difference that does **not** actually exist.  
- The probability of making a Type 1 error is equal to the **significance level (Œ±)**.  

### Type 2 Error (False Negative):  
- Occurs when the **null hypothesis (H‚ÇÄ)** is **not rejected**, even though it is actually false.  
- In simple terms: Failing to detect an effect or difference that **actually exists**.  
- The probability of making a Type 2 error is denoted by **Œ≤**.  

## Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

### Answer:  
In hypothesis testing, the choice between a **one-tailed** or **two-tailed** test depends on the direction of the research question.

### One-Tailed Test:  
- Tests for an effect in **one specific direction** (either greater than or less than).  
- The entire significance level (Œ±) is placed on one side of the distribution.  
- Used when the researcher wants to check for improvement or reduction only.  

**Example:**  
- Checking if a new drug **increases** recovery rate (right-tailed test).  
- Checking if a process **reduces** defects (left-tailed test).  

### Two-Tailed Test:  
- Tests for an effect in **both directions** (either greater or less).  
- The significance level (Œ±) is split equally between both tails of the distribution.  
- Used when the researcher wants to check for any difference, regardless of direction.  

**Example:**  
- Checking if a new teaching method results in **any change** in test scores (increase or decrease).  

## Q8. What is the Z-test, and when is it used in hypothesis testing?

### Answer:  
The **Z-test** is a statistical test used to determine whether there is a significant difference between sample and population means, or between the means of two samples, when the population standard deviation is known.

### When to Use a Z-test:  
- The sample size is **large** (usually n ‚â• 30).  
- The population follows a **normal distribution** or the sample size is large enough for the Central Limit Theorem to apply.  
- The **population standard deviation (œÉ)** is known.  

## Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

### Answer:  
The **Z-score** measures how many standard deviations a data point or sample mean is away from the population mean.

### Formula to Calculate Z-score:  

\[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
\]

Where:  
- **Z** = Z-score  
- **ùëãÃÑ** = Sample mean  
- **Œº** = Population mean  
- **œÉ** = Population standard deviation  
- **n** = Sample size  

### What it Represents:  
- The Z-score shows the position of the sample mean relative to the population mean.  
- A **high absolute Z-score** indicates that the sample mean is far from the population mean, suggesting the result may be statistically significant.  
- The Z-score is compared to critical values to decide whether to reject the null hypothesis.  

## Q10. What is the T-distribution, and when should it be used instead of the normal distribution?

### Answer:  
The **T-distribution**, also called **Student's T-distribution**, is a probability distribution similar to the normal distribution but with heavier tails. It is used when estimating population parameters with small sample sizes and when the population standard deviation is unknown.

### When to Use the T-distribution:  
- The sample size is **small** (n < 30).  
- The population standard deviation (œÉ) is **unknown**.  
- The data is approximately **normally distributed**.  

## Q11. What is the difference between a Z-test and a T-test?

### Answer:  
The **Z-test** and **T-test** are both used in hypothesis testing to compare means, but they differ based on sample size and whether the population standard deviation is known.

### Z-test:  
- Used when the **population standard deviation (œÉ)** is **known**.  
- Sample size is **large** (n ‚â• 30).  
- Assumes the population follows a **normal distribution**.  

### T-test:  
- Used when the **population standard deviation (œÉ)** is **unknown**.  
- Sample size is **small** (n < 30).  
- Based on the **T-distribution**, which has heavier tails to account for more uncertainty.  

## Q12. What is the T-test, and how is it used in hypothesis testing?

### Answer:  
The **T-test** is a statistical test used to determine whether there is a significant difference between the means of two groups or between a sample mean and a known value, especially when the sample size is small and the population standard deviation is unknown.

### When to Use the T-test:  
- Sample size is **small** (n < 30).  
- **Population standard deviation (œÉ)** is unknown.  
- The population is approximately **normally distributed**.  

### Types of T-tests:  
1. **One-sample T-test:**  
   Tests if the sample mean differs from a known population mean.  
2. **Independent two-sample T-test:**  
   Compares the means of two independent groups.  
3. **Paired T-test:**  
   Compares means from the same group at different times or under different conditions.  

### How it is Used:  
1. State the null and alternative hypotheses.  
2. Calculate the **T-statistic** using sample data.  
3. Compare the T-statistic to critical values from the **T-distribution table**, or use the **P-value**.  
4. Make a decision to reject or fail to reject the null hypothesis.  

## Q13. What is the relationship between Z-test and T-test in hypothesis testing?

### Answer:  
The **Z-test** and **T-test** are both statistical methods used to test hypotheses about population means. They serve a similar purpose but are used under different conditions.

### Relationship and Key Points:  
- Both tests help determine if there is a significant difference between sample data and population parameters.  
- The **T-test** is considered an extension of the **Z-test** for situations where the population standard deviation (œÉ) is unknown.  
- As the sample size increases, the **T-distribution** becomes similar to the **normal distribution**, making the **T-test and Z-test results nearly identical** for large samples.  

## Q14. What is a confidence interval, and how is it used to interpret statistical results?

### Answer:  
A **confidence interval (CI)** is a range of values, calculated from sample data, that is likely to contain the true population parameter (such as the mean or proportion) with a certain level of confidence.

### Example:  
A 95% confidence interval means we are 95% confident that the true population parameter lies within that range.

### How it is Used:  
- Confidence intervals give an estimate of the range within which the true value lies.  
- Instead of giving just a single estimate (point estimate), a CI provides a range, making the result more informative.  
- Wider intervals indicate more uncertainty; narrower intervals indicate more precision.  

### Interpretation:  
If the 95% CI for the difference between two means is (2.5, 5.0), we can say with 95% confidence that the true difference lies between 2.5 and 5.0.  
If the CI includes **0**, it suggests that there may be no significant difference.

## Q15. What is the margin of error, and how does it affect the confidence interval?

### Answer:  
The **margin of error (MoE)** is the amount added and subtracted from the sample estimate to create a confidence interval. It reflects the uncertainty or possible error in estimating the population parameter.

### Formula (for mean with known œÉ):  
\[
\text{Margin of Error} = Z \times \frac{\sigma}{\sqrt{n}}
\]  
Where:  
- **Z** = Z-score based on confidence level (e.g., 1.96 for 95%)  
- **œÉ** = Population standard deviation  
- **n** = Sample size  

### How it Affects the Confidence Interval:  
- The **margin of error** determines the width of the confidence interval.  
- A **larger margin of error** results in a **wider** confidence interval, indicating more uncertainty.  
- A **smaller margin of error** results in a **narrower** confidence interval, indicating more precision.  

## Q16. How is Bayes' Theorem used in statistics, and what is its significance?

### Answer:  
**Bayes' Theorem** is a fundamental concept in statistics that describes how to update the probability of a hypothesis based on new evidence or information.

### Formula:  
\[
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
\]  
Where:  
- **P(A|B)** = Probability of event A given event B has occurred (posterior probability)  
- **P(B|A)** = Probability of event B given event A has occurred (likelihood)  
- **P(A)** = Prior probability of event A  
- **P(B)** = Total probability of event B  

### How It Is Used:  
- To update probabilities as new information becomes available.  
- Common in **medical testing**, **machine learning**, **decision making**, and **spam detection**.  
- Helps in understanding the impact of new evidence on existing beliefs.  

### Significance:  
- It provides a systematic way to revise predictions or beliefs based on new data.  
- It plays a key role in **Bayesian statistics**, which is widely used in modern data analysis and real-world decision-making under uncertainty.

## Q18. What is the Chi-square goodness of fit test, and how is it applied?

### Answer:  
The **Chi-square goodness of fit test** is a statistical test used to determine whether a set of observed frequencies matches the expected frequencies for a categorical variable.

### Formula:  
\[
\chi^2 = \sum \frac{(O - E)^2}{E}
\]  
Where:  
- **O** = Observed frequency  
- **E** = Expected frequency  

### Steps to Apply the Test:  
1. State the null hypothesis (**H‚ÇÄ**): The observed frequencies fit the expected distribution.  
2. Calculate the expected frequencies.  
3. Compute the Chi-square statistic using the formula.  
4. Compare the calculated value to the critical value from the Chi-square distribution table based on degrees of freedom.  
5. Make a decision:  
   - If calculated œá¬≤ > critical value ‚Üí Reject H‚ÇÄ (significant difference).  
   - If calculated œá¬≤ ‚â§ critical value ‚Üí Fail to reject H‚ÇÄ (no significant difference).  

## Q19. What is the F-distribution, and when is it used in hypothesis testing?

### Answer:  
The **F-distribution** is a continuous probability distribution that arises when comparing two sample variances. It is commonly used in hypothesis testing to analyze whether there are significant differences between group variances or means.

### When it is Used:  
- **ANOVA (Analysis of Variance):** To compare means of three or more groups.  
- **Test for Equality of Variances:** To compare the variances of two populations.  

## Q20. What is an ANOVA test, and what are its assumptions?

### Answer:  
**ANOVA (Analysis of Variance)** is a statistical test used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others.

### Assumptions of ANOVA:  
1. **Independence:** The observations in each group are independent.  
2. **Normality:** The data within each group is approximately normally distributed.  
3. **Homogeneity of Variances (Equal Variances):** The variances across groups are roughly equal.  

## Q21. What are the different types of ANOVA tests?

### Answer:  
There are several types of **ANOVA (Analysis of Variance)** tests, each designed for specific experimental setups and research questions.

### 1. One-Way ANOVA:  
- Compares the means of **three or more independent groups** based on a **single independent variable**.  
- Example: Comparing test scores of students taught using three different teaching methods.

### 2. Two-Way ANOVA:  
- Examines the effects of **two independent variables** on a dependent variable.  
- Also tests for **interaction effects** between the two factors.  
- Example: Studying the impact of both teaching method and gender on student performance.

### 3. Repeated Measures ANOVA:  
- Used when the **same subjects** are measured multiple times under different conditions or over time.  
- Example: Measuring blood pressure of patients before, during, and after treatment.

### 4. Two-Way Repeated Measures ANOVA:  
- Combines the features of two-way ANOVA and repeated measures.  
- Involves **two within-subjects factors** or **one within-subjects and one between-subjects factor**.  

## Q22. What is the F-test, and how does it relate to hypothesis testing?

### Answer:  
The **F-test** is a statistical test used to compare the variances of two or more groups to determine if they are significantly different. It is often used as part of **hypothesis testing**, especially in tests like ANOVA.

### How it Relates to Hypothesis Testing:  
- The F-test helps determine whether observed differences in group variances or means are statistically significant.  
- It is based on the **F-distribution**, which is always positive and right-skewed.  
- The F-statistic is calculated as the ratio of two variances:  

\[
F = \frac{\text{Variance between groups}}{\text{Variance within groups}}
\]

## Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for discrete uniform distribution (e.g., integers from 1 to 6)
low = 1
high = 7  # upper bound is exclusive

# Generate a discrete uniform distribution
x = np.arange(low, high)
pmf = [1 / (high - low) for _ in x]  # Equal probability for all outcomes

# Plot the PMF
plt.stem(x, pmf, basefmt=" ")
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

## Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.


```python
from scipy.stats import bernoulli

# Function to calculate PDF of Bernoulli distribution
def bernoulli_pdf(p):
    x = [0, 1]  # Possible outcomes for Bernoulli distribution
    pdf_values = bernoulli.pmf(x, p)

    for xi, pdf in zip(x, pdf_values):
        print(f'P(X = {xi}) = {pdf:.2f}')

# Example: Bernoulli distribution with probability of success p = 0.6
bernoulli_pdf(0.6)
```

## Q15. Implement hypothesis testing using Z-statistics for a sample dataset.


```python
import numpy as np
from scipy.stats import norm

# Sample dataset (e.g., test scores)
data = np.array([78, 85, 80, 76, 79, 82, 84, 88, 75, 81])

# Population mean (hypothesized value)
mu_0 = 80

# Known population standard deviation (for Z-test)
sigma = 5

# Sample mean and size
sample_mean = np.mean(data)
n = len(data)

# Calculate Z-statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))
print(f"Z-statistic: {z_stat:.2f}")

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print(f"P-value: {p_value:.4f}")

# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

## Q16. Create a confidence interval for a dataset using Python and interpret the result.


```python
import numpy as np
from scipy.stats import norm

# Sample dataset (e.g., heights of individuals)
data = np.array([160, 165, 162, 170, 168, 164, 169, 167, 166, 171])

# Calculate sample mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
n = len(data)

# Confidence level (e.g., 95%)
confidence = 0.95
z_critical = norm.ppf(1 - (1 - confidence) / 2)

# Margin of error
margin_of_error = z_critical * (std_dev / np.sqrt(n))

# Confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

## Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.


```python
import numpy as np
from scipy.stats import norm

# Known population parameters
population_mean = 100
population_std = 15

# Sample data
sample = np.array([102, 98, 105, 100, 110, 95, 99, 101, 104, 103])
sample_mean = np.mean(sample)
sample_size = len(sample)

# Perform Z-test
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample mean and the population mean.")
```

## Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
sample = np.array([102, 98, 105, 100, 110, 95, 99, 101, 104, 103])
sample_mean = np.mean(sample)
sample_size = len(sample)

# Known population parameters
population_mean = 100
population_std = 5  # assumed known

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed test at 5% significance level
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"Z-critical value (¬±{z_critical:.2f}) for 95% confidence")

# Plot standard normal distribution with decision regions
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3)
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.title('Two-Tailed Z-Test with Decision Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Interpretation
if abs(z_score) > z_critical:
    print("Reject the null hypothesis: Sample mean significantly differs from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

## Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to visualize Type 1 and Type 2 errors
def visualize_errors(mu_null=0, mu_alt=1, sigma=1, alpha=0.05):
    # Critical value for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)

    # Define x range
    x = np.linspace(mu_null - 4*sigma, mu_alt + 4*sigma, 1000)

    # Null and Alternative distributions
    y_null = norm.pdf((x - mu_null) / sigma)
    y_alt = norm.pdf((x - mu_alt) / sigma)

    plt.plot(x, y_null, label='Null Hypothesis Distribution', color='blue')
    plt.plot(x, y_alt, label='Alternative Hypothesis Distribution', color='green')

    # Rejection regions
    critical_left = mu_null - z_critical * sigma
    critical_right = mu_null + z_critical * sigma

    # Shade Type 1 Error regions (Œ±)
    plt.fill_between(x, y_null, where=(x < critical_left), color='red', alpha=0.3, label='Type 1 Error (Œ±)')
    plt.fill_between(x, y_null, where=(x > critical_right), color='red', alpha=0.3)

    # Shade Type 2 Error region (Œ≤)
    plt.fill_between(x, y_alt, where=(x > critical_left) & (x < critical_right), color='orange', alpha=0.3, label='Type 2 Error (Œ≤)')

    plt.axvline(critical_left, color='black', linestyle='--', label='Critical Values')
    plt.axvline(critical_right, color='black', linestyle='--')

    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function
visualize_errors()
```

## Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.


```python
import numpy as np
from scipy.stats import t

# Function to calculate confidence interval for sample mean
def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)

    # t-critical value
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Simulate sample data
np.random.seed(4)
sample = np.random.normal(100, 15, 30)

# Calculate confidence interval
lower, upper = confidence_interval(sample)
print(f"95% Confidence Interval for the Sample Mean: ({lower:.2f}, {upper:.2f})")
```

## Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data.


```python
import numpy as np
from scipy.stats import t

# Function to calculate margin of error
def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_critical * std_err
    return moe

# Simulate sample data
np.random.seed(5)
sample = np.random.normal(80, 10, 40)

# Calculate margin of error
moe = margin_of_error(sample)
print(f"Margin of Error at 95% confidence level: {moe:.2f}")
```

## Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.


```python
# Example of Bayesian Inference using Bayes' Theorem

# Define probabilities
P_Disease = 0.01         # Prior probability: 1% chance a person has the disease
P_Positive_given_Disease = 0.99  # Test is 99% accurate if disease present (True Positive)
P_Positive_given_NoDisease = 0.05  # 5% False Positive rate
P_NoDisease = 1 - P_Disease

# Apply Bayes' Theorem
# P(Disease | Positive) = [P(Positive | Disease) * P(Disease)] / P(Positive)

# Total probability of testing positive
P_Positive = (P_Positive_given_Disease * P_Disease) + (P_Positive_given_NoDisease * P_NoDisease)

# Posterior probability: probability of having the disease given a positive test
P_Disease_given_Positive = (P_Positive_given_Disease * P_Disease) / P_Positive

# Display results
print(f"Probability of having the disease given a positive test: {P_Disease_given_Positive * 100:.2f}%")
```

## Q14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
